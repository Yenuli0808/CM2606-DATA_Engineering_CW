# **Data Profiling**

---

The dataset was profiled to understand its schema, data types, missing values and duplication levels prior to pipeline design.

```python
import pandas as pd
import numpy as np

# Set display options for better viewing
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)
```

```python
# Load the dataset from GitHub
url="https://raw.githubusercontent.com/Yenuli0808/CM2606-DATA_Engineering_CW/main/Dataset/superstore.csv"

# Load the data
df = pd.read_csv(url)
print("✅ Dataset loaded successfully!")
print("\n=== Data set overview ===")
print(f"Dataset shape: {df.shape}")
```

```python
# First look at the data
print("==== FIRST 5 ROWS ====\n")
display(df.head())
```

```python
print("\n==== Dataset Info ====\n")
df.info()
```

```python
# Check column names
print("Columns in dataset:")
print(df.columns)
```

```python
# check missing values
print("\nMissing values per column:")
print(df.isnull().sum())
```

```python
# Check missing values
missing_values = df.isnull().sum()

print("Missing values per column:")
print(missing_values[missing_values > 0])
```

```python
# Check duplicate rows
duplicate_count = df.duplicated().sum()

print("Number of duplicate rows:", duplicate_count)
```

```python
print("\n===== SUMMARY STATS =====")
df.describe(include="all")
```

```python
# Date Column Inspection
df['Order.Date'] = pd.to_datetime(df['Order.Date'])
print(df['Order.Date'].dtype)
```

**Summary**

* The dataset consists of 51,290 records and 27 attributes. Data profiling confirmed no missing values and no duplicate records. Date columns were converted to appropriate datetime formats to support temporal aggregation.
